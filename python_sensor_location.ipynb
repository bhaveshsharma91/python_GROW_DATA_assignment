```python
#Importing all the required Libraries
#Pandas is used to create and do operations onn dataframe
#Matplotlib will help to visualizatize the data

import pandas as pd
import matplotlib.pyplot as plt

#reading data in-memory and then storing it into a variable df.
df = pd.read_csv("GrowLocations.csv")

# It was observed that the lattitude and longitude data were interchanged that was restricting the visualization output. 
# Swapping Latitude and Longitude columns
df['Latitude'], df['Longitude'] = df['Longitude'], df['Latitude']

# Filter df and create a new dataframe that will only include values that are within the provided bounding boxe values. 
df = df[(df['Latitude'] >= 50.681) & (df['Latitude'] <= 57.985) & (df['Longitude'] >= -10.592) & (df['Longitude'] <= 1.6848)]


'''Loading the map ==> creating the subplots to include all the points ==>
liminting the area to be visvalized based on the bounding box values provided'''
base_uk_map = plt.imread("map7.png")
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(base_uk_map, extent=[-10.592, 1.6848, 50.681, 57.985])

#Ploting the sensor locations on the map
for index, row in df.iterrows():
    x, y = row['Longitude'], row['Latitude']
    ax.plot(x, y, marker='X', markersize=2, color='red')
    

#Including tittles and lables to the map.
plt.title('Sensor Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
